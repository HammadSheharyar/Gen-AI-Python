### Task 5

Create dictionary named <mark>europs</mark> from provided lists: <mark>countries</mark> and <mark>capitals</mark> using **for** loop

Hint: using enumerate() function with for loop

````python
# Definition of countries and capital
countries = ['spain', 'france', 'germany', 'norway']
capitals = ['madrid', 'paris', 'berlin', 'oslo']
````

````python
# write code here
europs = {}
for i, capital  in enumerate(capitals):
    europs[capital] = countries[i]

print(europs)
````

````python
# print europe
for capital, country in zip(capitals, countries):
    europs[capital] = country

print(europs)
````
